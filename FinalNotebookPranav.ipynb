```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
CaliHousing = pd.read_csv('california-housing-data.csv')
CaliHousing.head()
```

```python
x = CaliHousing[CaliHousing['median_house_value']<=500000] #remove house value outliers
```

```python
box = sns.boxplot(y='median_income',data=CaliHousing,color='#f98dc9')
box.set(ylabel='Median Income in tens of thousands')
```

```python
sns.scatterplot(data=x, x='median_income', y='median_house_value',hue='ocean_proximity',palette="cubehelix_r")
plt.title('Relation between Income and House Value')
```

```python
plt.style.use('classic')
scale_factor=3 #sclaes down the size of the circles to make it readable 
scaled_price=[scale_factor*x for x in CaliHousing['median_income']]
plt.scatter(CaliHousing['longitude'],CaliHousing['latitude'],s=scaled_price, alpha=0.2,color='#27fdf5')
plt.title('Relationship between Income and Geographic location')
```
