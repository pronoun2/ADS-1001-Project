```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
CaliHousing = pd.read_csv('california-housing-data.csv')
CaliHousing.head()
```

```python
sns.cubehelix_palette(as_cmap=True)
```

```python
sns.pairplot(data=CaliHousing, hue='ocean_proximity');
```

```python
sns.scatterplot(data=CaliHousing, x='median_house_value', y='total_bedrooms', hue='ocean_proximity')
plt.title('Relation between House Value and Total Bedrooms')
```

```python
sns.scatterplot(data=CaliHousing, x='median_income', y='total_bedrooms')
plt.title('Relation between Income and Total Bedrooms')
```

```python
sns.scatterplot(data=CaliHousing, x='total_rooms', y='total_bedrooms', hue='ocean_proximity')
plt.title('Relation between Total Rooms and Total Bedrooms')
```

```python
sns.scatterplot(data=CaliHousing, x='households', y='total_bedrooms', hue='ocean_proximity')
plt.title('Relation between Number of Households and Total Bedrooms')
```

```python
sns.boxplot(data=CaliHousing, y="total_bedrooms", x="ocean_proximity")
plt.title('Total Bedrooms spread and outliers');
```

```python
sns.scatterplot(data=CaliHousing, x='median_income', y='median_house_value',hue='ocean_proximity')
plt.title('Relation between Income and House Value')
```

```python
plt.style.use('classic')
scale_factor=3 #sclaes down the size of the circles to make it readable 
scaled_price=[scale_factor*x for x in CaliHousing['median_income']]
plt.scatter(CaliHousing['longitude'],CaliHousing['latitude'],s=scaled_price, alpha=0.2)
plt.title('Relationsip between Income and Geographic location')
```
